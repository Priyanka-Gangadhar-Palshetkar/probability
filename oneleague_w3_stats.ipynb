### Tests for Population Means

$$z = \frac{\text{estimate} - \text{hypothesized value}}{\text{standard deviation of the estimate}}$$
also known as 

$$z =\frac{\bar{x} - \mu_0}{\sigma/\sqrt{n}}$$
**EXAMPLE**

Consumption of sugar-sweetened beverages (SSB) has been positively associated with weight gain and obesity and negatively associated with the intake of important micronutrients.  One study used data from the data from the National Health and Nutrition Examination Survey to estimate SSB consumption among adolescents.  More than 2400 individuals provided data for this study.  The mean consumption was 298 calories per day.  

You survey 100 students at your large university and find the average consumption of SSB's per day to be 262 calories. Is there evidence that the average calories per day at your university differ from the U.S. average?

- State the null and alternative hypothesis
$$H_0 = \text{The students at TNS don't consume different from national average}$$
$$H_a = \text{The students at TNS are different from national average}$$
- Standard deviation is $\sigma = 435$
```{r}
z <- (262 - 298)/(435/sqrt(100))
z
```
```{r}
pnorm(z)
```


**EXAMPLE**(Two-sided)

Jeffrey, as an eight-year old, established a mean time of 16.43 seconds for swimming the 25-yard freestyle, with a standard deviation of 0.8 seconds. For the 15 swims, Jeffrey's mean time was 16 seconds. Frank thought that the goggles helped Jeffrey to swim faster than the 16.43 seconds. Conduct a hypothesis test using a preset $\alpha$ = 0.05.

```{r}
z <- (16.0 - 16.43)/(.8/sqrt(15))

```
